**Calculadora de índice de masa corpora**l
Crear un programa que pida al usuario su nombre, apellido paterno, apellido materno,edad, peso y estatura, desplegarlos en pantalla junto con su indice de masa corporal (IMC).
El programa no puede permitir que ningun dato quede vacio, además de asegurarse de que en los campos de edad, peso y estatura del usuario introduzca una cifra.
Todo esto antes de proceder con el calculo del IMC.
Siguiendo la formula:
Peso / estatura2
Finalmente, subir este proyecto a un repositorio de Github
IMC = P/E2


```python
def obtener_dato(mensaje, tipo=str, valida=lambda x: x.strip() != "" ):
  while True:
    dato = input(mensaje).strip()
    try:
      dato = tipo(dato)
      if valida(dato):
        return dato
    except ValueError:
      # pass
      print("Entrada inválida. Por favor, inténtelo de nuevo.")
```

```python
def calcular_imc(altura, peso):
  imc = peso / (altura**2)
  return imc
  # print(imc)
```

```python
# Obtener datos del usuario
nombre_completo = obtener_dato("Ingrese su nombre completo: ")
edad = obtener_dato("Ingrese su edad: ", int, lambda x: x > 18)
peso = obtener_dato("Ingrese su peso actual en kilogramos: ", float, lambda x: x > 0)
altura = obtener_dato("Ingrese su altura en metros: ", float, lambda x: 0 <= x <= 100)
```

```python
# Calcular imc
imc = calcular_imc(altura, peso)
```

```python
# Desplegar información
print("/Información del Usuario:")
print(f"Nombre completo: {nombre_completo}")
print(f"Edad: {edad} años")
print(f"Peso: {peso} kilogramos")
print(f"Altura: {altura} metros")
print(f"imc: {imc:.2f}")
```
