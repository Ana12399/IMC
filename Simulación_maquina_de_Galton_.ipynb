**Simulación de una máquina de Galton** de 3000 canicas. En total tendrá 12 niveles de obstáculos -deberás decidir si va a caer a un lado o al otro 12 veces. El resultado final será un histograma que represente la cantidad de canicas en cada contenedor, como el siguiente -No olvides colocar nombre a los ejes y un título al gráfico. Deberás emplear dos funciones, una para calcular los resultados de las canicas y la segunda para la graficación del histograma.


```python
import numpy as np
import matplotlib.pyplot as plt

def Simulación_Máquina_Galton(num_canicas = 3000, num_niveles = 12):
    resultado = []

    for i in range(num_canicas):
        position = 0
        for i in range(num_niveles):
            step = np.random.choice([-1,1])
            position += step
        resultado.append(position)
    print(resultado)
    return resultado


# Parámetros
num_canicas = 3000
num_niveles = 12

# simulacion
resultado = Simulación_Máquina_Galton(num_canicas, num_niveles)


def plot_histogram(resultado, num_niveles):
    plt.hist(resultado, bins=2*num_niveles+1, edgecolor='yellow')
    plt.title('SIMULACIÓN MÁQUINA DE GALTON')
    plt.xlabel('CONTENEDOR')
    plt.ylabel('NUMERO DE CANICAS')
    plt.grid(True)
    plt.show()

# Gráfica del histograma
plot_histogram(resultado, num_niveles)
```
